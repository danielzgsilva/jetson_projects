```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
cwd = os.getcwd()
log_dir = os.path.join(cwd, 'power_logs')

baseline_log_file = 'baseline.log'
compress_1_0_log_file = 'compress_1.0.log'
compress_0_8_log_file = 'compress_0.8.log'
compress_0_6_log_file = 'compress_0.6.log'
compress_0_4_log_file = 'compress_0.4.log'
compress_0_2_log_file = 'compress_0.2.log'
compress_0_1_log_file = 'compress_0.1.log'
```

```python
baseline_logs = pd.read_csv(os.path.join(log_dir, baseline_log_file), names=['Current', 'Power'], header=0)
compress_1_0_logs = pd.read_csv(os.path.join(log_dir, compress_1_0_log_file), names=['Current', 'Power'], header=0)
compress_0_8_logs = pd.read_csv(os.path.join(log_dir, compress_0_8_log_file), names=['Current', 'Power'], header=0)
compress_0_6_logs = pd.read_csv(os.path.join(log_dir, compress_0_6_log_file), names=['Current', 'Power'], header=0)
compress_0_4_logs = pd.read_csv(os.path.join(log_dir, compress_0_4_log_file), names=['Current', 'Power'], header=0)
compress_0_2_logs = pd.read_csv(os.path.join(log_dir, compress_0_2_log_file), names=['Current', 'Power'], header=0)
compress_0_1_logs = pd.read_csv(os.path.join(log_dir, compress_0_1_log_file), names=['Current', 'Power'], header=0)
```

```python
logs = [baseline_logs, compress_1_0_logs, compress_0_8_logs, 
        compress_0_6_logs, compress_0_4_logs, compress_0_2_logs, compress_0_1_logs]

names = ['Baseline', '1.0', '0.8', '0.6', '0.4', '0.2', '0.1']
```

```python
start = 48
scale = 1000
end_offset = 1
```

```python
fig = plt.figure()
for log_file in logs:
    n = len(log_file)
    plt.plot(log_file['Power'][start:n-end_offset] / scale)
    
plt.legend(names)
plt.title('Power in watts over time during inference', fontsize=14)
plt.xlabel('Seconds', fontsize=12)
plt.ylabel('Watts', fontsize=12)
plt.show()
```

```python
fig = plt.figure()
for log_file in logs:
    n = len(log_file)
    plt.plot(log_file['Current'][start:n-end_offset] / scale)
    
plt.legend(names)
plt.title('Current in amps over time during inference', fontsize=14)
plt.xlabel('Seconds', fontsize=12)
plt.ylabel('Amps', fontsize=12)
plt.show()
```
